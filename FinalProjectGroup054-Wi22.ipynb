## Background & Prior Work

### Background

From the lightbulb and internet [1] to football huddles [2] and Morse code [3], deaf people have been an influential part of American society since its origins. Public sentiment towards the Deaf community regarding their rights to access and language has been controversial and polarizing. Before we get into the importance of positive language and representation, we must first define some terminology regarding the Deaf community and American Sign Language. You may have noticed that we have been capitalizing the “D” in deaf. As with many cultures, the Deaf community takes pride in their abilities and perseverance in the face of ableism. To establish their status as a cultural group, the deaf and hard-of-hearing(hoh) use “D” deaf when discussing their community and its members. “d” deaf is used to denote the medical/scientific analysis regarding deafness, as well as, a general disconnect from the Deaf community [4]. Within the Deaf community there is the term “Deaf Gain" which “accentuates the benefits of being deaf” [5] instead of focusing on the loss of hearing. One of these gains is the ability to communicate in a loud space through the use of signed language. American Sign Language(ASL) is a form of physically expressed communication used by the Deaf community since the early 1800’s. ASL became officially recognized as a language by the academic community when William Stokoe published Dictionary of American Sign Language on Linguistic Principles in 1965 [6]. The use of ASL has long been regarded by the Deaf community as essential for growth and communication [7], though this same sentiment has not always been the viewpoint of American society. The term “audism” refers to the act of discrimination and separation on the basis of one's ability to hear[8]. Examples of audism include, but are not limited to: refusing to use written communication, prohibiting the use of ASL, refusal of service on the basis of deafness, refusal to hire an ASL interpreter, and not learning to sign for a deaf offspring. 
    
Throughout American history, the deaf and hard of hearing have been labeled many things, such as: deaf and dumb, deaf mute, and hearing impaired. All of these terms carry negative connotations in regards to a person's intellect, speaking ability, and the implication that they could “be better” if they could hear typically. To counteract these implications, the Deaf community now prefers the terms deaf and hard-of-hearing, which refer to their inability to conduct and process sound. It must also be mentioned that the term “mute” is now considered a choice in the Deaf community and not a reflection of one's ability to vocalize. Through their sense of comradery and pride, the Deaf community have been able to experience and accomplish many great strides in terms of disability rights and representation. For example, the passage of the Americans with Disabilities Act (ADA) has been said to have been largely influenced by the media exposure given to the Deaf students of Gallaudet University in 1988 during their Deaf President Now protest [9]. Due to the trail blazed by actress Marlee Matlin, Deaf performers like Nyle Dimarco, Millicent Simmonds from _A Quiet Place_(2018), and Lauren Ridolff in _The Eternals_(2021) have all been hired to portray Deaf characters in a more positive and representative light. Until recently, deaf characters were often portrayed by hearing performers as exemplified by Hawkeye in _The Avengers_(2012) and the suspense/thriller _Hush_(2016). 

Whether they are viewed as disabled or an inspiration, there’s no denying the impact the deaf have on American culture. Having proper representation and positive sentiment portrayed to the general public has proven to be beneficial to the accessibility and awareness of the Deaf community. In order to continue along this trajectory of inclusion we decided it would be beneficial to analyze the media rhetoric regarding the Deaf community to gain a better understanding of how representation has changed over time. 

### Prior Work

Media platforms have demonstrated both positive and unconscious bias in the Deaf community. The sentiment of language carries both negative and positive interpretations through journalistic outlets. This expression of negative opinion can be reflected through fact-based news that can create discrimination toward underrrepresented groups without that intent. 

Deeper into this research, there are studies that are taking the first step to reduce unconscious bias against people with disabilities by focusing on the language being used in mainstream media. 

In one study, researchers developed a system that can neutralize ableism in texts such as articles, books and speeches. By doing so, they developed an algorithm that identifies any subjective text and generates new sentence pairs to avoid linguistic ableism [10]. For example, if the algorithm reads “Apparently, she’s confined to a wheelchair,” the invective word would be replaced with the word “uses” to neutralize the statement. Their language database contains 1500 sentence pairs that are from movies scripts, articles, and speech transcripts that are later utilized in a BERT encoder [10]. This NLP system supports automatic identification and alternative writing against disability in text. 

It is important to mention, there is no standard framework to identify ableist language, but this study explores the possibility of using a pre-training concurrent model using Wiki Neutrality Corpus (WNC) from the very first generative model that can neutralize biased text. No parallel corpus of alesit language currently exists [10]. However, this study proposes a new framework to include more phrases and words that are non-inclusive of disability by collecting sub categorical text that can be used through albelm or audism language. This study also used human evaluation in their model’s generated sentences to maximize the accuracy and neutrality. Human evaluation with their BERT encoder has helped detect hidden words or phrases that are subcategories of ableism in the media.

# Data Cleaning

The data that we scraped is messy, and so we want to create a function to help us clean our data and get it into a standardized format. We want this function to do a number of things.

#### Define a dataframe cleaning function that:
- Renames the columns
- Drops null values
- Renames the source strings to their acronyms
- Converts the date strings to pd.datetime format
- Translate unicode artifacts from the text using unicodedata.normalize

We want the columns to be renamed for easier access later in the notebook, and similarly we want the news sources to be renamed to their acronyms for similar ease of use. We want to drop null values so that we aren't working with any incomplete data. We want to convert our date strings into a format that is easier to work with. And finally we want to translate any unicode artifacts that are left over from scraping the articles off of a website into their plaintext equivalents.

```python
def df_clean(df, source):
    df = df[['headline','date','source','url','text']]
    
    df.dropna(inplace=True)
        
    df['source'] = source

    df['date'] = pd.to_datetime(df['date'])
    df = df[~(df['date']<='1980-01-01')]
    
    df['text'] = df['text'].apply(lambda t: unicodedata.normalize('NFKD', t))
        
    return df
```

## Generating Sentiment Analysis Scores

In this next section, we are creating new dataframes which will have the analysis results in addition to defining a few helper functions for our analysis. We will be using TextBlob to find the sentiment scores of the different articles. Textblob will return two variables summarizing the sentiment. First, it will return a polarity score (-1 to 1), where -1 is a negative sentiment and +1 is a positive sentiment. It also returns a subjectivity score (0 to 1) which scores how opinionated a text is, where 0 is the least subjective and 1 is the most subjective.

Define a function that finds sentiment within a given text:

```python
def get_sentiment(text):
    blob = TextBlob(text)
    polarity, subjectivity = blob.sentiment
    return polarity, subjectivity
```

Now we will apply this function to each news dataframe and add two columns with the polarity score (ranging from -1 to 1) and subjectivity score (ranging from 0 to 1).

```python
combined_df[['polarity', 'subjectivity']] = combined_df.apply(lambda x: get_sentiment(x['text']),
                                                              axis=1, result_type='expand')
```

Generally, we can see an slight upward trend in polarity, indicating more positive language being used over the last two decades. Let's statistically analyze this trend. 

First, we need to determine whether our data is gaussian or not. This will determine the types of tests we can run on it. In addition to plotting our actualy data, we will plot a curve showing the expected distribution, under the hypothesis that it is normally distributed. 

```python
#make a subset for the years after 2000
subset = combined_df[combined_df['year'] >=2000].copy()
plt.figure(figsize=(10, 6))
xs = np.arange(subset['polarity'].min(), subset['polarity'].max(), 0.01)
fit = stats.norm.pdf(xs, np.mean(subset['polarity']), np.std(subset['polarity']))
plt.plot(xs, fit, label='Normal Dist.', lw=3)
plt.hist(subset['polarity'], 50, density=True, label='Actual Data', color='pink');
plt.legend();
plt.xlabel("polarity", fontsize = 20)
```

This graph shows the actuat data in pink, and the blue line is the expected data density, with the same mean and standard deviation, if the data is normal. We can see a pretty close fit between the two. However, we will also run a normal test to more accurately asses whether our data accepts the null hypothesis and follows a normal distribution. 

```python
stat, pval = stats.normaltest(subset['polarity'])
alpha = 0.05
print('\nNormaltest p-value is: {:1.2e} \n'.format(pval))
if pval < alpha:
    print('We have evidence to reject the null hypothesis.')
else:
    print('We do not have evidence to reject the null hypothesis. This is not a normal distribution')
```

This p-value is extremely below 0.05, indicating this does NOT follow a gaussian distribution. One thing we can keep in mind is the accuracy of this particular test on large datasets (ours is close to 9,500 articles), which has shown to give lower p-values as the data set grows bigger, which means for large datasets it tends to always report a value which rejects the null. 

However, to be safe, we will assume our data is not gaussian and carry on with non-parametric tests. First, we will perform a Spearman correlational test, which reports a p-value and a correlational coefficient (-1 being a strong negative correlation and 1 being a strong positive correlation).

```python
coef, p = stats.spearmanr(subset['polarity'],subset['year'])
print('Spearmans correlation coefficient: %.3f' % coef)
print('\n p-value is: {:1.2e} \n'.format(p))
```

We can see a slight positive correlation, with a very small p-value indicating that this slight positive correlation is not due to chance. 

To further analyze this data, we will perform a Kruskal-Wallis test. This is a non-parametric alterantive to an ANOVA test, in which it will test the null hypothesis that the median of all groups are equal. In our case, we will be splitting our data into subsets of 3-4 years and compare the median polarity between the groups. 

```python
first = subset[(subset['year'] >=2000) & (subset['year'] < 2003) ].copy()
second = subset[(subset['year'] >=2003) & (subset['year'] < 2006) ].copy()
third  = subset[(subset['year'] >=2006) & (subset['year'] < 2009) ].copy()
fourth = subset[(subset['year'] >=2009) & (subset['year'] < 2012) ].copy()
fifth = subset[(subset['year'] >=2008) & (subset['year'] < 2010) ].copy()
sixth  = subset[(subset['year'] >=2012) & (subset['year'] < 2015) ].copy()
seventh = subset[(subset['year'] >=2015) & (subset['year'] < 2018) ].copy()
eighth = subset[(subset['year'] >=2018) & (subset['year'] < 2022) ].copy()
hstat, pval = stats.kruskal(first['polarity'], second['polarity'],third['polarity'],fourth['polarity'],fifth['polarity'],sixth['polarity'],seventh['polarity'],eighth['polarity'])
print('Kruskal Wallis coefficient: %.3f' % hstat)
print('\n p-value is: {:1.2e} \n'.format(pval))
```

Our results show a very high test-statistic of 54.062. The higher the value for this test-statistic, the larger of a difference between the groups. Our p-value is very low, indicating that this is not due to random chance. Given this information, we can assume a difference of sentiment over time.

## TF-IDF Exploration

### Top Keywords

We can use TF-IDF techniques to extrapolate keywords from a text. We can see which words are the most impactful to given documents.

We will be dividing our articles up into documents based on their news source and year they were published in.

```python
# Dividing up our text data into combined strings by source and year
documents = defaultdict(lambda: defaultdict(str))

for _, _, _, source, _, text, _, _, year in combined_df.itertuples():
    documents[source][year] += text
```

We will be using the sklearn library to process these documents into their TF-IDFs, and need them in a list format.

Here we convert our dictionary into a list of strings, and preserve the keys of the dictionary in a separate list called document_names

```python
corpus = []
document_names = []
for source in documents.keys():
    for year in documents[source].keys():
        corpus.append(documents[source][year])
        document_names.append((source, year))
```

```python
vectorizer = TfidfVectorizer(stop_words=stopwords.words('english'))
X = vectorizer.fit_transform(corpus)
words = list(vectorizer.get_feature_names_out())
```

Now that we have our TF-IDFs in X, we want to look to see what the most significant words in each document are.

Here we find the top 10 keywords for each document and add them into a new dictionary: top_tfidfs.

```python
top_tfidfs = defaultdict(lambda: defaultdict(list))
for tfidfs, (source, year) in zip(X.toarray(), document_names):
    sorted_tfidfs = tfidfs.argsort()[:-11:-1]
    
    for index in sorted_tfidfs:
        top_tfidfs[source][year].append(words[index])
```

```python
top_tfidfs['guard'][2021]
```

```python
# Getting a list of all years present in our dataset
all_years = set()
for source in top_tfidfs.keys():
    all_years = all_years.union(set(top_tfidfs[source].keys()))
all_years = list(all_years)
all_years.sort()
```

Now that we have the top keywords for each source for each year, we can print them out. We decided to only print the top 5 keywords of each source in each year to keep the representation readable.

```python
for year in all_years:
    print(f'\n--- Year {year} ---')
    for source in top_tfidfs.keys():
        if year in top_tfidfs[source]:
            print(f'{source}: {top_tfidfs[source][year][:5]}')
```

### Importance of Keywords

We looked at the top keywords for entire years, but we might also want to track how significant different keywords are for all articles.

For this we're going to have each article be a separate document in our corpus. We will then be able to graph the TF-IDF scores for different keywords over time.

```python
corpus = []
for text in combined_df['text']:
    corpus.append(text)
```

```python
vectorizer = TfidfVectorizer(stop_words=stopwords.words('english'))
X = vectorizer.fit_transform(corpus)
```

X is a sparse matrix where every document in our corpus is a row, and every column represents a word. There is a column for every different word in the corpus (around 150,000) and the value in a given column represents the TF-IDF value of the word for the given article. This is equivalent to the significance of the keyword in the article. If the keyword isn't present in an article the TF-IDF will just be 0.

We start by getting the column index of different keywords we want to track

```python
words = list(vectorizer.get_feature_names_out())

disabled_index = words.index('disabled')
culture_index = words.index('culture')
deaf_index = words.index('deaf')
cochlear_index = words.index('cochlear')
handicapped_index = words.index('handicapped')
art_index = words.index('art')
```

We can then take those columns and add them back into our dataframe

```python
combined_df['disabled'] = X[:, disabled_index].toarray()
combined_df['culture'] = X[:, culture_index].toarray()
combined_df['deaf'] = X[:, deaf_index].toarray()
combined_df['cochlear'] = X[:, cochlear_index].toarray()
combined_df['handicapped'] = X[:, handicapped_index].toarray()
combined_df['art'] = X[:, art_index].toarray()
```

We can now graph the distribution of significance for each keyword
